
&emsp;&emsp;线性分类器通常无法直接训练非线性可分问题, 不过可以采用一定的特征转换来将优化目标转化为一个线性可分来实现。 比如, 我们可以对异或门的两个特征计算平方差作为新的特征, 即$ x_3 = (x_1  - x_2)^2$, 这样问题就变成了一维线性可分的。 当然, 面对更加复杂的问题, 我们通常是不知道什么样的特征是可以让问题变得线性可分的, 尽管一些先验知识和直觉能够帮助我们去找到这样的特征, 但这是一件很麻烦的事情。那么, 怎样才能够让模型自己去学习这样的一些组合特征, 从而实现对非线性可分问题的准确分类呢? 下期将为大家实现包含多个神经元及其层数的多层感知机(MLP)
